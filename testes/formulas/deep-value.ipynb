```python
import pandas as pd

# https://sistemaswebb3-listados.b3.com.br/indexPage/day/IBOV?language=pt-br
dataColumns =  pd.read_csv('../dados/IBOVDia_04-08-22.csv', sep=';', skiprows=1, encoding='latin-1')
columns = dataColumns.columns
columns
```

```python
data =  pd.read_csv('../dados/IBOVDia_04-08-22.csv', sep=';', skiprows=2, skipfooter=2, header=None, encoding='latin-1', engine='python')
data = data.rename(columns={0: columns[0], 1: columns[1], 2: columns[2], 3: columns[3], 4: columns[4]})
data.drop([5], axis=1)
```

```python
cod = data['Código'].to_numpy()
```

```python
dataI =  pd.read_csv('../dados/statusinvest-busca-avancada1.csv', sep=';',encoding='latin-1', engine='python')
#Margem Ebit 0.01
#Liquidez média diária 1.000.000
dataI.shape
```

```python
dataIBOV=dataI[dataI["TICKER"].isin(cod)]
columns = dataIBOV.columns
columns
```

```python
dataIBOV.shape
```

```python
pd.options.display.max_columns = None
dataIBOV
```

```python
def corrigir_nomes(nome):
    nome = nome.replace('.', '')
    return nome
```

```python
dataIBOV['EV/EBIT'] = dataIBOV['EV/EBIT'].apply(corrigir_nomes)
dataIBOV = dataIBOV.replace({',': '.'}, regex=True)
pd.options.display.max_columns = None
#pd.options.display.max_rows = None
dataIBOV
```

```python
dataIBOV = dataIBOV.astype({"EV/EBIT":'float'})
dataIBOV = dataIBOV.sort_values(by=["EV/EBIT"])

dataIBOV.head(30)
```

```python
cut = round(len(dataIBOV.index)/2)
dataIBOV.head(cut).to_csv("deep-value.csv")
```
