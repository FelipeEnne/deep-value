```python
data =  pd.read_csv('IBOVDia_04-08-22.csv', sep=';', skiprows=2, skipfooter=2, header=None, encoding='latin-1', engine='python')
data = data.rename(columns={0: columns[0], 1: columns[1], 2: columns[2], 3: columns[3], 4: columns[4]})
data.drop([5], axis=1)
```

```python
cod = data['Código'].to_numpy()
cod
```

```python
dataI =  pd.read_csv('statusinvest-busca-avancada.csv', sep=';',encoding='latin-1', engine='python')
#P/L 0.1
#P/VP 0.1
#VPA 0.1
#LPA 0.1
#Liquidez média diária 1.000.000
dataI.shape
```

```python
dataI["TICKER"].isin(cod)
```

```python
dataIBOV=dataI[dataI["TICKER"].isin(cod)]
columns = dataIBOV.columns
columns
```

```python
dataIBOV.shape
```

```python
dataIBOV
```

```python
dataIBOV = dataIBOV.replace({',': '.'}, regex=True)
dataIBOV = dataIBOV.astype({columns[26]:'float', columns[27]:'float', columns[1]:'float'})
dataIBOV
```

```python
dataIBOV["formula-gharant"] = (dataIBOV[columns[26]]*dataIBOV[columns[27]]*22.5)**0.5
dataIBOV
```

```python
dataIBOV["margem-seguranca"] = (dataIBOV["formula-gharant"]/dataIBOV["PRECO"]-1)*100
dataIBOV
```

```python
dataIBOV = dataIBOV.sort_values(by=["margem-seguranca"], ascending=False)
dataIBOV.head(30)
```

```python
cut = round(len(dataIBOV.index)/2)
dataIBOV.head(cut).to_csv("formula-gharant.csv")
```
