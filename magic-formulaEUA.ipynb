```python
import pandas as pd

pd.options.display.max_columns = None
#pd.options.display.max_rows = None

csvBA = 'statusinvest-busca-avancadaEUA.csv'
#https://statusinvest.com.br/acoes/eua/busca-avancada
#Liquidez média diária 200.000
#EV/EBIT positivo
```

```python
data =  pd.read_csv(csvBA, sep=';',encoding='latin-1', engine='python')
data.shape
```

```python
data = data.dropna(subset=['EV/EBIT', 'MARGEM EBIT'])
data
data
```

```python
def fixNames(name):
    name = name.replace('.', '')
    return name
```

```python
data = data.astype({"EV/EBIT":'string'})
data = data.astype({"MARGEM EBIT":'string'})

data['EV/EBIT'] = data['EV/EBIT'].apply(fixNames)
data['MARGEM EBIT'] = data['MARGEM EBIT'].apply(fixNames)
data = data.replace({',': '.'}, regex=True)

data = data.astype({"EV/EBIT":'float'})
data = data.astype({"MARGEM EBIT":'float'})

data
```

```python
data = data.sort_values(by=["EV/EBIT"])
data.head(30)
```

```python
data["point"] = 1
data = data.sort_values(by=["EV/EBIT"])
data["point EV/EBIT"] = data['point'].cumsum()
data = data.sort_values(by=["MARGEM EBIT"], ascending=False)
data["point MARGEM EBIT"] = data['point'].cumsum()
data["Magic Stock"] = data["point EV/EBIT"] + data["point MARGEM EBIT"]
data = data.sort_values(by=["Magic Stock"])

data.head(30)
```
