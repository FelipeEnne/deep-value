```python
#EV/EBIT 0,01
#Liquidez média diária 250.000,00
#Valor de mercado  250.000.000,00

data =  pd.read_csv('statusinvest-busca-avancada-2023-01-06.csv', sep=';',encoding='latin-1')
columns = data.columns
columns
```

```python
cod = data['TICKER'].to_numpy()
len(cod)
```

```python
data = data.sort_values(by=["EV/EBIT"])
data.head(10)
```

```python
data["point"] = 1
data["pointEV/EBIT"] = data['point'].cumsum()
```

```python
data = data.sort_values(by=["ROIC"], ascending=False)
data["pointROIC"] = data['point'].cumsum()
```

```python
data["pointFinal"] = data["pointEV/EBIT"] + data["pointROIC"]
```

```python
data = data.sort_values(by=["pointFinal"])
```
