```python
import pandas as pd

pd.options.display.max_columns = None

csvBA = 'statusinvest-busca-avancadaEUA.csv'
#https://statusinvest.com.br/acoes/eua/busca-avancada
#Liquidez média diária 200.000
#EV/EBIT positivo
```

```python
data =  pd.read_csv(csvBA, sep=';',encoding='latin-1', engine='python')
data.shape
```

```python
data = data.dropna(subset=['ROA', 'P/L'])
data
data
```

```python
def fixNames(name):
    name = name.replace('.', '')
    return name
```

```python
data = data.astype({"ROA":'string'})
data = data.astype({"P/L":'string'})

data['ROA'] = data['ROA'].apply(fixNames)
data['P/L'] = data['P/L'].apply(fixNames)
data = data.replace({',': '.'}, regex=True)

data = data.astype({"ROA":'float'})
data = data.astype({"P/L":'float'})

data
```

```python
#data = data.sort_values(by=["ROA"], ascending=False)
data = data.sort_values(by=["ROA"])
data.head(30)
```

```python
data = data.sort_values(by=["P/L"], ascending=False)
#data = data.sort_values(by=["P/L"])
data.head(30)
```

```python
data["point"] = 1
data = data.sort_values(by=["P/L"])
data["point P/L"] = data['point'].cumsum()
data = data.sort_values(by=["ROA"], ascending=False)
data["point ROA"] = data['point'].cumsum()
data["Magic Stock"] = data["point P/L"] + data["point ROA"]
data = data.sort_values(by=["Magic Stock"])

pd.options.display.max_rows = None

data.head(100)
```
