```python
import pandas as pd

dataFG =  pd.read_csv("formula-gharant.csv")
dataDV =  pd.read_csv("deep-value.csv")
pd.options.display.max_columns = None
```

```python
cod = dataDV['TICKER'].to_numpy()
dataF=dataFG[dataFG["TICKER"].isin(cod)]
dataF
```

```python
dataF.shape
```

```python
dataF = dataF.sort_values(by=["EV/EBIT"])
dataF.head(10)
```

```python
dataF["point"] = 1
```

```python
dataF["pointEV/EBIT"] = dataF['point'].cumsum()
```

```python
dataF = dataF.sort_values(by=["ROIC"], ascending=False)
dataF["pointROIC"] = dataF['point'].cumsum()
```

```python
dataF["pointFinal"] = dataF["pointEV/EBIT"] + dataF["pointROIC"]
```

```python
dataF = dataF.sort_values(by=["pointFinal"])
dataF20 = dataF.head(20)
```

```python
dataF20
```

```python
sumMargin = dataF20["margem-seguranca"].sum()
sumMargin
```

```python
dataF20['invest'] =  dataF20["margem-seguranca"]/sumMargin
dataF20 = dataF20.sort_values(by=["invest"], ascending=False)
dataF20
```

```python
invest = 200
dataF20['investStock'] =  round(dataF20["invest"]*invest/dataF20["PRECO"])
dataF20
```

```python
dataF20.to_csv("final-05-08-2022.csv")
```
